```python
import pandas as pd
import numpy as np
```

```python
steam_df = pd.read_csv('steam_reviews.csv')
steam_df.head()
```

```python
# Drop columns
drop_col = ['funny', 
            'helpful', 
            'is_early_access_review']
steam_df.drop(drop_col, inplace=True, axis=1)
```

```python
steam_df.head()
```

```python
steam_df.count()
```

```python
steam_df = steam_df.rename(columns={"review": "text"})
steam_df.head()
```

```python
steam_df['target'] = np.where(steam_df['recommendation']!= 'Not Recommended', True, False)
steam_df.head(20)
```

```python
steam_df[["target"]] *=1
steam_df.head(15)
```

```python
steam_df.dropna(axis=0, how="any", inplace=True)
steam_df.head(50)
```

```python
steam_df.count()
```

```python
steam_df.tail(50)
```

```python
steam_df.to_csv("clean_steam_reviews.csv")
```
